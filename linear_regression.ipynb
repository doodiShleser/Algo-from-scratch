**implemention of linear regression algorithm**

```python
import numpy as np
from numpy import random as rnd
from sklearn.metrics import r2_score
```

```python
def g_noise(n):
    return rnd.normal(0,0.1,n)

def polynom(x,coef):
    p=0
    for i,c in enumerate(coef):
        p = p+c*pow(x,i)
    return p
```

**simple linear regression**

```python
floats = rnd.ranf(10)
x_1 = floats.reshape(10,1)
y_1 = 2*floats+g_noise(10)
inv = np.linalg.inv(np.matmul(x_1.transpose(),x_1))
h1 = np.matmul(inv, np.matmul(x_1.transpose(),y_1))
r2 = r2_score(h1*floats,y_1)
```

```python
import matplotlib.pyplot as plt
axis = np.linspace(0,1,num=10)
plt.figure()
plt.scatter(x_1,y_1)
plt.plot(axis,h1*axis,'r')
plt.title('R^2 = '+str(round(r2,2)))
plt.show()
```

**linear regression with independent coef**

```python
y_2 = 4+3*floats+g_noise(10)
ones = np.ones(10)
x_2 = np.column_stack((x_1,ones))
inv = np.linalg.inv(np.matmul(x_2.transpose(),x_2))
h2 = np.matmul(inv, np.matmul(x_2.transpose(),y_2))
r2 = r2_score(h2[1] +h2[0]*floats,y_2)
```

```python

plt.figure()
plt.scatter(x_1,y_2)
plt.plot(axis,h2[0]*axis+h2[1],'r')
plt.title('R^2 = '+str(round(r2,2)))
plt.show()
```

**quadratic regression**

```python
coef = rnd.randint(10,size=3)
x = np.dot(0.005,range(1000))  #to be in range [0,5]
p20 = rnd.choice(x,20)
y_3 = polynom(p20,coef)+g_noise(20)
x_3 = np.column_stack((np.ones(20),p20,p20**2))
inv = np.linalg.inv(np.matmul(x_3.transpose(),x_3))
h3 = np.matmul(inv, np.matmul(x_3.transpose(),y_3))
axis = np.linspace(0,5,num=20)
r2 = r2_score(polynom(p20,[h3[0],h3[1],h3[2]]),y_3)
```

```python
new = np.linspace(0,5,num=20)
plt.figure()
plt.scatter(p20,y_3)
plt.plot(axis,polynom(axis,[h3[0],h3[1],h3[2]]),'r')
plt.title('R^2 = '+str(round(r2,2)))
plt.show()
```
